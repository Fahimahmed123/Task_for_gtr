```python

import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

#  Final Fixed Trading Strategy Class
class GoldenCrossTrading:
    """
    A class to implement golden cross trading strategy - COMPLETE FIXED VERSION
    """

    def __init__(self, symbol, from_date, to_date, budget=5000):
        """
        Initialize the trading strategy
        """
        self.symbol = symbol
        self.from_date = from_date
        self.to_date = to_date
        self.budget = budget
        self.initial_budget = budget
        self.data = None
        self.position = False
        self.shares_owned = 0
        self.trades = []
        self.portfolio_values = []

    def download_data(self):

        print(f"Downloading {self.symbol} data from {self.from_date} to {self.to_date}...")
        try:
            self.data = yf.download(self.symbol, start=self.from_date, end=self.to_date, progress=False)
            print(f"Downloaded {len(self.data)} trading days")
            return True
        except Exception as e:
            print(f"Error: {e}")
            return False

    def clean_data(self):
        """Clean the data by removing duplicates and handling NaN values"""
        print("\n Cleaning data...")

        # Remove duplicates
        before = len(self.data)
        self.data = self.data[~self.data.index.duplicated(keep='first')]

        # Handle NaN values
        nan_before = self.data.isna().sum().sum()
        self.data = self.data.ffill().bfill()

        print(f" Removed {before - len(self.data)} duplicates")
        print(f" Filled {nan_before} NaN values")

    def calculate_moving_averages(self):
        """Calculate 50-day and 200-day moving averages"""
        print("\nCalculating moving averages...")
        self.data['MA50'] = self.data['Close'].rolling(window=50).mean()
        self.data['MA200'] = self.data['Close'].rolling(window=200).mean()
        print("50-day and 200-day MAs calculated")

    def generate_signals(self):
        """Generate trading signals based on golden cross"""
        print("\nGenerating trading signals...")

        # Initialize signal column
        self.data['Signal'] = 0

        # Get values as arrays
        ma50 = self.data['MA50'].values
        ma200 = self.data['MA200'].values

        # Create position array
        position = np.zeros(len(self.data))
        for i in range(len(self.data)):
            if not np.isnan(ma50[i]) and not np.isnan(ma200[i]):
                if ma50[i] > ma200[i]:
                    position[i] = 1

        # Find crossovers
        for i in range(1, len(position)):
            if position[i] == 1 and position[i-1] == 0:
                self.data.iloc[i, self.data.columns.get_loc('Signal')] = 1
            elif position[i] == 0 and position[i-1] == 1:
                self.data.iloc[i, self.data.columns.get_loc('Signal')] = -1

        buy_signals = len(self.data[self.data['Signal'] == 1])
        sell_signals = len(self.data[self.data['Signal'] == -1])
        print(f" Found {buy_signals} BUY signals and {sell_signals} SELL signals")

    def execute_trades(self):
        """Execute trading strategy based on signals"""
        print("\nExecuting trades...")

        # Reset trading state
        self.position = False
        self.shares_owned = 0
        self.budget = self.initial_budget
        self.trades = []
        self.portfolio_values = []

        # Convert to lists properly
        dates = list(self.data.index)
        signals = list(self.data['Signal'].values)
        closes = list(self.data['Close'].values)

        for i in range(len(dates)):
            date = dates[i]
            signal = signals[i]
            current_price = float(closes[i])

            # Track portfolio value
            if self.position:
                current_value = self.budget + (self.shares_owned * current_price)
            else:
                current_value = self.budget

            self.portfolio_values.append({
                'Date': date,
                'Value': current_value,
                'Price': current_price
            })

            # Check for BUY signal
            if signal == 1 and not self.position:
                max_shares = int(self.budget / current_price)
                if max_shares > 0:
                    cost = max_shares * current_price
                    self.shares_owned = max_shares
                    self.budget -= cost
                    self.position = True

                    self.trades.append({
                        'date': date,
                        'type': 'BUY',
                        'price': current_price,
                        'shares': max_shares,
                        'value': cost
                    })
                    print(f"  BUY {max_shares} shares @ ${current_price:.2f} on {date.strftime('%Y-%m-%d')}")

            # Check for SELL signal
            elif signal == -1 and self.position:
                revenue = self.shares_owned * current_price
                self.budget += revenue

                self.trades.append({
                    'date': date,
                    'type': 'SELL',
                    'price': current_price,
                    'shares': self.shares_owned,
                    'value': revenue
                })
                print(f"  SELL {self.shares_owned} shares @ ${current_price:.2f} on {date.strftime('%Y-%m-%d')}")

                self.shares_owned = 0
                self.position = False

        # Force close position on last day
        if self.position:
            last_date = dates[-1]
            last_price = float(closes[-1])
            revenue = self.shares_owned * last_price
            self.budget += revenue

            self.trades.append({
                'date': last_date,
                'type': 'SELL (FORCED)',
                'price': last_price,
                'shares': self.shares_owned,
                'value': revenue
            })
            print(f"   FORCED SELL {self.shares_owned} shares @ ${last_price:.2f} on {last_date.strftime('%Y-%m-%d')}")

            self.shares_owned = 0
            self.position = False

        print(f"\n Completed {len(self.trades)} trades")

    def evaluate_performance(self):
        """Calculate and display performance metrics"""
        print("\n" + "="*60)
        print(" PERFORMANCE EVALUATION")
        print("="*60)

        # Calculate returns
        total_profit = self.budget - self.initial_budget
        roi = (total_profit / self.initial_budget) * 100 if self.initial_budget > 0 else 0

        # Calculate buy & hold
        first_price = float(self.data.iloc[0]['Close'])
        last_price = float(self.data.iloc[-1]['Close'])
        buy_hold_shares = int(self.initial_budget / first_price)
        buy_hold_value = buy_hold_shares * last_price
        buy_hold_profit = buy_hold_value - self.initial_budget
        buy_hold_roi = (buy_hold_profit / self.initial_budget) * 100 if self.initial_budget > 0 else 0

        # Display results
        print(f"\n Strategy Performance:")
        print(f"  Initial Budget: ${self.initial_budget:,.2f}")
        print(f"  Final Budget:   ${self.budget:,.2f}")
        print(f"  Total Profit:   ${total_profit:,.2f}")
        print(f"  ROI:            {roi:.2f}%")
        print(f"  Total Trades:   {len(self.trades)}")

        print(f"\n Buy & Hold Benchmark:")
        print(f"  Initial Investment: ${self.initial_budget:,.2f}")
        print(f"  Shares Purchased:   {buy_hold_shares}")
        print(f"  Final Value:        ${buy_hold_value:,.2f}")
        print(f"  Profit:             ${buy_hold_profit:,.2f}")
        print(f"  ROI:                {buy_hold_roi:.2f}%")

        # Calculate win rate
        if len(self.trades) >= 2:
            buy_trades = [t for t in self.trades if t['type'] == 'BUY']
            sell_trades = [t for t in self.trades if t['type'] in ['SELL', 'SELL (FORCED)']]

            if len(buy_trades) > 0 and len(sell_trades) > 0:
                profits = []
                for i in range(min(len(buy_trades), len(sell_trades))):
                    profit = sell_trades[i]['value'] - buy_trades[i]['value']
                    profits.append(profit)

                if profits:
                    winning_trades = len([p for p in profits if p > 0])
                    win_rate = (winning_trades / len(profits)) * 100
                    print(f"\n Trade Statistics:")
                    print(f"  Win Rate: {win_rate:.2f}% ({winning_trades}/{len(profits)})")
                    print(f"  Avg Profit per Trade: ${np.mean(profits):,.2f}")

        # Compare
        if roi > buy_hold_roi:
            print(f"\ Strategy outperformed Buy & Hold by {roi - buy_hold_roi:.2f}%")
        elif roi < buy_hold_roi:
            print(f"\n Strategy underperformed Buy & Hold by {buy_hold_roi - roi:.2f}%")
        else:
            print(f"\ Strategy performed equally to Buy & Hold")

    def plot_results(self):
        """Plot trading results - ULTIMATE FIXED VERSION"""
        fig, axes = plt.subplots(3, 1, figsize=(15, 10))

        # Plot 1: Price and MAs
        ax1 = axes[0]
        ax1.plot(self.data.index, self.data['Close'], label='Close Price', alpha=0.7, linewidth=1, color='blue')
        ax1.plot(self.data.index, self.data['MA50'], label='50-day MA', linewidth=2, color='orange')
        ax1.plot(self.data.index, self.data['MA200'], label='200-day MA', linewidth=2, color='red')

        # Mark signals
        buy_mask = self.data['Signal'] == 1
        sell_mask = self.data['Signal'] == -1

        buy_dates = self.data.index[buy_mask]
        buy_prices = self.data.loc[buy_mask, 'Close']
        sell_dates = self.data.index[sell_mask]
        sell_prices = self.data.loc[sell_mask, 'Close']

        if len(buy_dates) > 0:
            ax1.scatter(buy_dates, buy_prices, color='green', marker='^', s=100, label='Buy', zorder=5)
        if len(sell_dates) > 0:
            ax1.scatter(sell_dates, sell_prices, color='red', marker='v', s=100, label='Sell', zorder=5)

        ax1.set_title(f'{self.symbol} - Golden Cross Strategy', fontsize=14, fontweight='bold')
        ax1.set_ylabel('Price ($)')
        ax1.legend(loc='best')
        ax1.grid(True, alpha=0.3)

        # Plot 2: Volume - FIXED VERSION
        ax2 = axes[1]
        # Convert to numpy array and ensure it's 1-dimensional
        volume_data = self.data['Volume'].values.flatten()
        ax2.bar(range(len(volume_data)), volume_data, alpha=0.5, color='blue', width=0.8)
        ax2.set_title('Trading Volume', fontsize=12)
        ax2.set_ylabel('Volume')
        ax2.set_xticks([])  # Remove x-tick labels for volume plot
        ax2.grid(True, alpha=0.3)

        # Plot 3: Portfolio
        ax3 = axes[2]
        if self.portfolio_values:
            portfolio_df = pd.DataFrame(self.portfolio_values)
            portfolio_df.set_index('Date', inplace=True)
            ax3.plot(portfolio_df.index, portfolio_df['Value'], color='purple', linewidth=2, label='Portfolio Value')
            ax3.axhline(y=self.initial_budget, color='red', linestyle='--', alpha=0.5, label='Initial Budget')
            ax3.set_title('Portfolio Value Over Time', fontsize=12)
            ax3.set_xlabel('Date')
            ax3.set_ylabel('Value ($)')
            ax3.legend(loc='best')
            ax3.grid(True, alpha=0.3)

        plt.tight_layout()
        plt.show()

    def print_trades(self):
        """Print trade history"""
        if not self.trades:
            print("No trades executed")
            return

        print("\n" + "="*60)
        print(" TRADE HISTORY")
        print("="*60)

        # Calculate summary statistics
        total_buy_value = sum(t['value'] for t in self.trades if 'BUY' in t['type'])
        total_sell_value = sum(t['value'] for t in self.trades if 'SELL' in t['type'])

        for i, trade in enumerate(self.trades, 1):
            date_str = trade['date'].strftime('%Y-%m-%d')
            if 'BUY' in trade['type']:
                print(f"{i:2}. {date_str} | {trade['type']:12} | "
                      f"Price: ${trade['price']:.2f} | Shares: {trade['shares']:3} | "
                      f"Value: ${trade['value']:.2f}")
            else:
                print(f"{i:2}. {date_str} | {trade['type']:12} | "
                      f"Price: ${trade['price']:.2f} | Shares: {trade['shares']:3} | "
                      f"Value: ${trade['value']:.2f}")

        print(f"\n Trade Summary:")
        print(f"  Total Buy Value:  ${total_buy_value:,.2f}")
        print(f"  Total Sell Value: ${total_sell_value:,.2f}")
        print(f"  Net Profit:       ${total_sell_value - total_buy_value:,.2f}")

    def run(self):
        """Execute complete trading strategy"""
        print("\n" + "="*60)
        print(f"GOLDEN CROSS TRADING STRATEGY FOR {self.symbol}")
        print("="*60)

        if not self.download_data():
            return False

        self.clean_data()
        self.calculate_moving_averages()
        self.generate_signals()
        self.execute_trades()
        self.evaluate_performance()
        self.plot_results()
        self.print_trades()

        return True

#  Run for Multiple Stocks
print("\n" + "="*60)
print("STARTING TRADING STRATEGY SIMULATION")
print("="*60)

# Test with different stocks
stocks = ['AAPL', 'MSFT', 'GOOGL']

for symbol in stocks:
    trader = GoldenCrossTrading(symbol, "2018-01-01", "2023-12-31", budget=5000)
    trader.run()
    print("\n" + "="*60 + "\n")

#  Performance Comparison
print("\n" + "="*60)
print("PERFORMANCE COMPARISON TABLE")
print("="*60)

results = []
for symbol in stocks:
    trader = GoldenCrossTrading(symbol, "2018-01-01", "2023-12-31", budget=5000)
    trader.download_data()
    trader.clean_data()
    trader.calculate_moving_averages()
    trader.generate_signals()
    trader.execute_trades()

    strategy_profit = trader.budget - trader.initial_budget
    strategy_roi = (strategy_profit / trader.initial_budget) * 100

    # Buy & hold
    first_price = float(trader.data.iloc[0]['Close'])
    last_price = float(trader.data.iloc[-1]['Close'])
    buy_hold_shares = int(trader.initial_budget / first_price)
    buy_hold_value = buy_hold_shares * last_price
    buy_hold_roi = ((buy_hold_value - trader.initial_budget) / trader.initial_budget) * 100

    results.append({
        'Symbol': symbol,
        'Strategy ROI': f"{strategy_roi:.2f}%",
        'Buy & Hold ROI': f"{buy_hold_roi:.2f}%",
        'Difference': f"{strategy_roi - buy_hold_roi:.2f}%",
        'Trades': len(trader.trades)
    })

results_df = pd.DataFrame(results)
print("\n Comparison Results:")
print(results_df.to_string(index=False))

# Cell 5: Interactive Test Function
def test_stock(symbol, from_date="2018-01-01", to_date="2023-12-31", budget=5000):
    """
    Test the strategy on any stock symbol
    """
    print(f"\n Testing strategy on {symbol}...")
    trader = GoldenCrossTrading(symbol, from_date, to_date, budget)
    trader.run()
    return trader

print("\n You can test any stock using: test_stock('SYMBOL')")
print("   Examples: test_stock('TSLA'), test_stock('AMZN'), test_stock('NVDA')")

# Test with Tesla

test_stock('TSLA')
```
